````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df_prevrenda = pd.read_csv('01.1 - Profissao- Cientista de Dados_M10_support material.csv')
df_prevrenda.head()
````

````python
df_prevrenda.info()
````

````python
df_prevrenda['posse_de_veiculo'].value_counts()
````

````python
10082+4918
````

````python
df1 = df_prevrenda[['posse_de_veiculo', 'posse_de_imovel','renda']]
df1
````

````python
ax= sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df1, dodge=True, ci = 90)

plt.xticks([0, 1],['False = sem_veiculo', 'True = com_veiculo'])
````

**Análise**: a posse de veículo trata-se de variável qualitativa que discrimina a renda de forma estatisticamente significante.

````python
ax= sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df1, dodge=True, ci = 90)

plt.xticks([0, 1],['False = sem_imovel', 'True = com_imovel'])
````

**Análise**: a posse de imóvel trata-se de variável qualitativa que discrimina a renda de forma estatisticamente significante.
