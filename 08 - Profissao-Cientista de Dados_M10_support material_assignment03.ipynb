1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df_prevrenda = pd.read_csv('01.1 - Profissao- Cientista de Dados_M10_support material.csv')
df_prevrenda
````

````python
df_prevrenda.info()
````

````python
df_prevrenda['data_ref'] = pd.to_datetime(df_prevrenda['data_ref'])
# convertendo a coluna 'data_ref'object para datetime
df1 = df_prevrenda
````

````python
tab_freq = pd.crosstab(df1['data_ref'],df1['sexo'])
tab = tab_freq.div(tab_freq.sum(axis = 1), axis = 0)
ax = tab.plot.bar(stacked = True, figsize=(10, 4))

tick_labs = df1['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(df1['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=45)


plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)

ax.set_title('Distribuição da variável Sexo no Tempo')
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
tick_labs = df1['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
````

````python
plt.rc('figure', figsize=(15, 25))
fig, axes = plt.subplots(4, 2)

sns.pointplot(ax=axes[0,0], x="data_ref", y="renda", hue = 'sexo',
                   data=df1, dodge=True, ci = 95)
sns.pointplot(ax=axes[0,1], x="data_ref", y="renda", hue = 'posse_de_veiculo',
                   data=df1, dodge=True, ci = 95)
sns.pointplot(ax=axes[1,0], x="data_ref", y="renda", hue = 'posse_de_imovel',
                   data=df1, dodge=True, ci = 95)
sns.pointplot(ax=axes[1,1], x="data_ref", y="renda", hue = 'tipo_renda',
                   data=df1, dodge=True, ci = 50)
sns.pointplot(ax=axes[2,0], x="data_ref", y="renda", hue = 'educacao',
                   data=df1, dodge=True, ci = 50)
sns.pointplot(ax=axes[2,1], x="data_ref", y="renda", hue = 'estado_civil',
                   data=df1, dodge=True, ci = 50)
sns.pointplot(ax=axes[3,0], x="data_ref", y="renda", hue = 'tipo_residencia',
                   data=df1, dodge=True, ci = 25)

axes[0,0].set_xticklabels(tick_labs, rotation=45)
axes[0,1].set_xticklabels(tick_labs, rotation=45)
axes[1,0].set_xticklabels(tick_labs, rotation=45)
axes[1,1].set_xticklabels(tick_labs, rotation=45)
axes[2,0].set_xticklabels(tick_labs, rotation=45)
axes[2,1].set_xticklabels(tick_labs, rotation=45)
axes[3,0].set_xticklabels(tick_labs, rotation=45)

axes[0,0].set_title('Média da Variável Resposta no Tempo por Sexo')
axes[0,1].set_title('Média da Variável Resposta no Tempo por Posse de Veículo')
axes[1,0].set_title('Média da Variável Resposta no Tempo por Posse de Imóvel')
axes[1,1].set_title('Média da Variável Resposta no Tempo por Tipo de Renda')
axes[2,0].set_title('Média da Variável Resposta no Tempo por Educação')
axes[2,1].set_title('Média da Variável Resposta no Tempo por Estado Civil')
axes[3,0].set_title('Média da Variável Resposta no Tempo por Tipo de Residência')

plt.subplots_adjust(wspace=0.15, hspace=0.50)
````

**Análise**: De acordo com os gráficos das médias da variáveis, verificou-se que a mais estáveis são as que teve como base a posse de veículo e o sexo, somentes estas servindo para a utilização de uma modelagem.
