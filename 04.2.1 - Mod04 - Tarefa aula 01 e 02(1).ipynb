#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na tempreatura global, A NASA disponibiliza dados de diferenças de de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. |ano|anomalia termica|
|:-:|:----:|
| 1900 | -0.08 |
| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
import pandas as pd
import numpy as np

# Seu código aqui

dados= [-0.08, -0.27, 0.12, -0.03, 0.26, 0.40, 1.02]

alteraçãotemp_sem_index= pd.Series(data=dados)

alteraçãotemp_sem_index
```
