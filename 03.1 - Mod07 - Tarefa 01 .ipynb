````python
# crie um df com os nomes de cada variável e o tipo de dados de cada variável
````

````python
metadados=pd.DataFrame(df.dtypes, columns=['tipo'])
metadados
````

````python
# Adicione uma coluna nesse dataframe chamada "qtd_categorias" e 
# coloque nela o número de categorias correspondente de cada variável.

metadados['qtd_categorias']=0

for var in metadados.index:
    metadados.loc[var,'qtd_categorias']=len(df.groupby([var]).size())

metadados
````

````python
# Crie variáveis dummy para as variáveis necessárias
# (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}
````

````python
print(df['sexo'].unique())
print(df['posse_de_veiculo'].unique())
print(df['posse_de_imovel'].unique())
print(df['tipo_renda'].unique())
print(df['educacao'].unique())
print(df['estado_civil'].unique())
print(df['tipo_residencia'].unique())
````

````python
df['sexo_masculino'] = pd.get_dummies(df['sexo'],drop_first=True) # Feminino = 0 e Masculino = 1
df['veiculo_yes'] = pd.get_dummies(df['posse_de_veiculo'],drop_first=True) # sem_veículo = 0 e com_veiculo = 1
df ['imovel_yes'] = pd.get_dummies(df['posse_de_veiculo'],drop_first=True) # sem_imóvel = 0 e com_imóvel = 1

dummies_renda=pd.get_dummies(df['tipo_renda'],drop_first=False) # variáveis: working, commercil associate, state servant, pensioner, student.
df2=pd.concat([df, dummies_renda], axis=1)

dummies_educ=pd.get_dummies(df['educacao'],drop_first=False) # 5 variáveis
df3=pd.concat([df2, dummies_educ], axis=1)


dummies_estciv=pd.get_dummies(df['estado_civil'],drop_first=False) # 4 variáveis
df4=pd.concat([df3, dummies_estciv], axis=1)


dummies_tresid=pd.get_dummies(df['tipo_residencia'],drop_first=False) # 6 variáveis
df5=pd.concat([df4, dummies_tresid], axis=1)
df5.head()
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# as variáveis dummy foram criadas no item anterior.
# Crie um dataframe apenas com as variáveis apropriadas para entrada no scikitlearn.

df_dummies = df5[['sexo_masculino', 'veiculo_yes', 'imovel_yes', 'Commercial associate',
       'Pensioner', 'State servant', 'Student', 'Working', 'Academic degree',
       'Higher education', 'Incomplete higher', 'Lower secondary',
       'Secondary / secondary special', 'Civil marriage', 'Married',
       'Separated', 'Single / not married', 'Widow', 'Co-op apartment',
       'House / apartment', 'Municipal apartment', 'Office apartment',
       'Rented apartment', 'With parents','idade',
       'tempo_emprego', 'possui_celular', 'possui_fone_comercial',
       'possui_fone', 'possui_email', 'qt_pessoas_residencia', 'mau']]
df_dummies.head()
````

````python
df_scikitlearn = df_dummies.rename({'Commercial associate': 'renda_comercial_associate',
       'Pensioner': 'renda_pensioner', 'State servant':'renda_state_servant',
       'Student':'renda_student', 'Working':'renda_working','Academic degree':'educ_academic_degree',
       'Higher education':'educ_higher_education', 'Incomplete higher':'educ_incomplete_higher',
       'Lower secondary':'educ_lower_secondary','Secondary / secondary special':'educ_secondary',
       'Civil marriage':'estciv_civil_marriage', 'Married':'estciv_married',
       'Separated':'estciv_separated', 'Single / not married':'estciv_single', 'Widow':'estciv_widow',
        'Co-op apartment':'resid_co-op_apartment','House / apartment':'resid_house_apartment',
        'Municipal apartment':'resid_municipal_apartment', 'Office apartment':'resid_office_apartment',
        'Rented apartment':'resid_rented_apartment', 'With parents':'resid_with_parents'},
        axis=1)

df_scikitlearn.head()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
freq2=pd.crosstab(df_scikitlearn['possui_email'], df_scikitlearn['veiculo_yes'])

freq2

                                 
````

Conclusão: a variável 'posse_de_veículo' é a mais poderosa para prever a probabilidade de mau=1, visto que a condição de possuir veículo apresenta maior numeros de eventos do que a condição de possuir e-mail.
