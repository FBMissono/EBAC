As 5 diferenças são:

### 2 - Acesse o link Scikit-learn - adaboost, leia a explicação (traduza se for preciso) e crie um jupyter notebook contendo o exemplo do Adaboost.

```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
from sklearn.ensemble import AdaBoostClassifier
```

```python
X, y = load_iris(return_X_y=True)
```

```python
clf = AdaBoostClassifier(n_estimators=100)
scores = cross_val_score(clf, X, y, cv=5)
```

### 3 - Cite 5 hiperparâmetros importantes no Adaboost. ###

**objeto base_estimator**: estimador base a partir do qual o conjunto impulsionado é contruído. É necessário suporte para ponderação de amostra, bem como classes_ e n_classes, atributos adequados. Se None (default), então o estimador base é DecisionTreeClassifier inicializado com max_depth=1.

**n_estimators**:o número máximo de estimadores em que o reforço é encerrado. Em caso de ajuste perfeito, o processo de aprendizagem é interrompido precocemente. Os valores devem estar no intervalo (1, infinito). N_estimators padrão é igual a 50.

**learning_rate**:peso aplicado a cada classificador em cada interação de reforço. Uma taxa de aprendizado mais alta aumenta a contribuição de cada classificador (OBS: o peso aumenta na medida em que o classificador apresenta maiores erros de predição). Há um trade-off entre os parâmetros learning_rate e n_estimators. Os valores devem estar no intervalo (0,0, infinito). 

**algorithm**:Se "SAMME.R", usa-se o algoritmo de reforço real, a base_estimator deve suportar o cálculo de probabilidades de classes. Se "SAMME", usa-se o algoritmo de reforço discreto SAMME.O algoritmo SAMME.R normalmente converge mais rápido que o SAMME, alcançando um erro de teste menor com menos interações de reforço. 

**random_state**: controla a semente aleatória dada em cada uma base_estimator em cada interação de reforço.Assim, ele só é usado quando base_estimator expõe arquivo random_state. Passe um int para saída reproduzível em várias chamadas de função.

### 4 - (Opcional) Utilize o GripSearch para encontrar os melhores hiperparâmetros para o conjunto de dados do exemplo (load_iris). ###

```python
list(range(1, 1001, 100))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
%%time
rf = AdaBoostClassifier()

params = {
    'n_estimators':list(range(1,1001, 100)),
    'learning_rate': [1],
    'random_state':[0]
}

grid_rf = GridSearchCV(estimator = rf, 
                      param_grid = params, 
                      scoring = 'accuracy', 
                      cv = 5)

grid_rf.fit(X, y)
```

```python
grid_rf.cv_results_['params']
```

```python
print(grid_rf.best_estimator_)
```

```python
print(grid_rf.score(X, y))
```
