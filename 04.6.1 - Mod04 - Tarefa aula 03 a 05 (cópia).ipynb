## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui

Vestibular=pd.DataFrame(df_mat)
Vestibular['Mat_Percent']=Vestibular['Qt_acertos']/24
Vestibular['Por_Qt_acertos']=df_por
Vestibular['Por_Percent']=Vestibular['Por_Qt_acertos']/18
Vestibular['Geo_Qt_acertos']=df_geo
Vestibular['Geo_Percent']=Vestibular['Geo_Qt_acertos']/8
Vestibular['Ing_Qt_acertos']=df_ing
Vestibular['Ing_Percent']=Vestibular['Ing_Qt_acertos']/8
Vestibular['His_Qt_acertos']=df_his
Vestibular['His_Percent']=Vestibular['His_Qt_acertos']/8
Vestibular['Fis_Qt_acertos']=df_fis
Vestibular['Fis_Percent']=Vestibular['Fis_Qt_acertos']/12
Vestibular['Qui_Qt_acertos']=df_qui
Vestibular['Qui_Percent']=Vestibular['Qui_Qt_acertos']/12

Vestibular_percent= Vestibular[['Mat_Percent','Por_Percent','Geo_Percent','Ing_Percent','His_Percent',
                                'Fis_Percent','Qui_Percent']]

Vestibular_percent
````

````python
# 2) Seu código aqui

Vestibular['Total_de_acertos'] = (
                    Vestibular['Qt_acertos']+Vestibular['Por_Qt_acertos'] + Vestibular['Geo_Qt_acertos']+
                    Vestibular['Ing_Qt_acertos']+Vestibular['His_Qt_acertos']+Vestibular['Fis_Qt_acertos']+
                    Vestibular['Qui_Qt_acertos']
                    )

Vestibular['Total_de_acertos']
````

````python
# 3) Seu código aqui

Vestibular

Vestibular['Total_em_percento']=(Vestibular['Total_de_acertos']/np.sum([24,18,8,8,8,12,12]))

Vestibular['Total_em_percento']
````

````python
# 4) Seu código aqui

vetor_bool=Vestibular['Total_de_acertos']>45

vetor_bool.sum()
````

````python
# 3)Nesse DataFrame faça uma contagem do campo vacina_categoria_nome.

vacAcre_copy['vacina_categoria_nome'].unique()
````

````python
# 4)a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista
# com esses valores.

vacAcre_copy.loc[:,'estabelecimento_valor':'estabelecimento_uf'].columns.unique()
````
